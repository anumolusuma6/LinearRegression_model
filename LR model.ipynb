```python
import pandas as pd
import numpy as np
df=pd.read_csv(r'C:\Users\dell\ML project\Linear Regression - Sheet1.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
#SPLITING DATA INTO DEPENDENT AND INDEPENDENT

X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=43)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
#PLOTTTING GRAPHH FOR TRAIN SET
#to visualize how well the model fits the training data.
#the reason why we selected scatter plot is because these points might have outliers like you can see now so if you do line graph, the line should also connect the outliers and it looks messy

plt.scatter(x_train,y_train , color="red")
plt.plot(x_train, lr.predict(x_train), color="blue")
plt.title('Y VS X (training set)')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
#PLOTTING GRAPH FOR TEST SET

plt.scatter(x_test,y_test,color="red")
plt.plot(x_test,y_pred,color="blue")
plt.title('Y VS X (test set)')
plt.xlabel('X')
plt.ylabel('Y')

#or
#plt.plot(x_test,lr.predict(x_test),color="blue")
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
